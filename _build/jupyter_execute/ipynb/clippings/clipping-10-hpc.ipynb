## Supercomputadores

- Definição não exata: SCs são computadores com extrema capacidade operacional para resolver problemas complexos e processar grandes quantidades de dados de maneira rápida.

- Desempenho comumente medido por FLOPS;
- Features:
    - Poder de processamento (FLOPS);
    - Componentes (comunicação, transferência);
    - Memória (capacidade de RAM de tera a petabytes);
    - Escalabilidade (computação distribuída/paralela);
    - Hardware (aceleração por GPUs, FPGAs etc.)
    - Consumo energético (resfriamento e uso)


- Motivações: 
    - Aceleração da descoberta científica;
    - Vantagem competitiva global;
    - Resolução de problemas complexos em variados domínios;


### Exaescala

- Iniciativas tais como o [exascaleproject](https://www.exascaleproject.org/about/) buscam a computação em exaescala.
- A exaescala busca poder computacional em escala de EXAFLOPS: 10<sup>18</sup> FLOPS.
