# Localização e refinamento de raízes

## Estudo de caso: salto do paraquedista

```python
%matplotlib inline 
import numpy as np
import matplotlib.pyplot as plt 

# parametros 
t = 12.0
v = 42.0
m = 70.0
g = 9.81
```

```python
# localizacao
a,b = 1,20
c = np.linspace(a,b,20)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f,'k--o');
plt.grid(True)
```

```python
# refinamento
a,b = 10,20
c = np.linspace(a,b,100)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f)
plt.plot(c,0*c,'r--')
plt.grid()
```

```python
# refinamento
a,b = 14,16
c = np.linspace(a,b,100)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f)
plt.plot(c,0*c,'r--')
plt.grid()
```

```python
# refinamento
a,b = 14.75,15.5
c = np.linspace(a,b,100)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f)
plt.plot(c,0*c,'r--')
plt.grid()
```

```python
# refinamento
a,b = 15.1,15.2
c = np.linspace(a,b,100)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f)
plt.plot(c,0*c,'r--')
plt.grid()
```

```python
# refinamento
a,b = 15.12,15.14
c = np.linspace(a,b,100)
f = g*m/c*(1 - np.exp(-c/m*t)) - v

plt.plot(c,f)
plt.plot(c,0*c,'r--')
plt.axvline(x=15.1278)
plt.axvline(x=15.1275)
plt.grid()
```

## Métodos iterativos

A primeira estrutura de controle fundamental para computação numérica é o laço `for`. Vamos ver como usá-lo para computar somatórias e produtórios. 

Por exemplo, vamos computar: 

$s = \sum_{i=1}^n i = i + i + \ldots + i = s_1 + s_2 + \ldots + s_{n}$, para um valor $n$ finito.

Podemos fazer isto da seguinte maneira:

```python
n = 10

s = 0.0
for i in range(n):
    s += i
```

Detalhando... Considere o par $(i,s_i)$. Há soma abaixo?

```python
s = 0.
for i in range(n):
    print(i+1,s)
```

```python
# Aqui há um incremento constante de 1
s = 0.
for i in range(n):
    s = s + 1
    print(i+1,s)
```

```python
s = 0.
for i in range(n):
    s += i # produz o mesmo que s <- s + i
    print(i+1,s)
```
